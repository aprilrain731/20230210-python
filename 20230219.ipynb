### 練習

抓取`年度國內主要金融指標`下來做分析

https://apiservice.mol.gov.tw/OdService/download/A17000000J-030244-dpp

````python
import requests
````

````python
url = "https://apiservice.mol.gov.tw/OdService/download/A17000000J-030244-dpp"
````

````python
data = requests.get(url).json()
````

````python
data
````

### 列出各年度的M2貨幣供給額

````python
# 檢視data的資料型態

type(data)
````

````python
# 檢視一下data有幾筆資料

len(data)
````

````python
# 檢視一下 data 的第一筆資料
data[0]
````

####方法一

````python
for i in data:
  print(i['年度'] ,'==>', i['貨幣供給額期底年增率-M2'])
 
````

####方法二

````python
# 列出data有幾項
for i in range(len(data)):
  print(i)
````

````python
# 以index形式
for i in range(len(data)):
  print(data[i]['年度'], "==>", data[i]['貨幣供給額期底年增率-M2'])
````
